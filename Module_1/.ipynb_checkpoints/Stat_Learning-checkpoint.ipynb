```python
m = plt.hist(data[data['diagnosis'] == 'M'].radius_mean, bins = 30, fc = (1, 0, 0, 0.5), label = 'Malignant')
b = plt.hist(data[data['diagnosis'] == 'B'].radius_mean, bins = 30, fc = (0, 1, 0, 0.5), label = 'Benign')
plt.legend()
plt.xlabel('Radius Mean Values')
plt.ylabel('Frequency')
plt.title('Histogram of Radius Mean for Benign and Malignant Tumors')
plt.show()

frequent_malignant_radius_mean = m[0].max()
index_frequent_malignant_radius_mean = list(m[0]).index(frequent_malignant_radius_mean)
most_frequent_malignant_radius_mean = m[1][index_frequent_malignant_radius_mean]
print("Most frequent malignant radius mean is: ",most_frequent_malignant_radius_mean)
```

## Outliers

```python
data_bening = data[data['diagnosis'] == 'B']
data_malignant = data[data['diagnosis'] == 'M']
desc = data_benign.radius_mean.describe()
Q1 = desc[4]
Q3 = desc[6]
IQR = Q3 - Q1
lower_limit = Q1 - IQR*1.5
upper_limit = Q3 + IQR*1.5
print("Anything outside this range is an outlier: (", lower_limit ,",", upper_limit,")")
data_bening[data_bening.radius_mean < lower_limit].radius_mean
print("Outliers: ",data_bening[(data_bening.radius_mean < lower_limit) | (data_bening.radius_mean > upper_limit)].radius_mean.values)
```

```python
df_no_outliers = data_bening[(data_bening.radius_mean >= lower_limit) & (data_bening.radius_mean <= upper_limit)]
```
